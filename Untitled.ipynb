# 概念

## 导数

1. 本质

   自变量增量极小时，对应的函数增量与自变量增量的比值的极限

2. 定义

   组成：函数：$f(x),\ x \in I$，自变量：$x_0$，自变量增量：$\Delta x$，函数值增量：$\Delta y$

   条件：$\displaystyle \lim _{\Delta x \rightarrow 0} \frac{\Delta y}{\Delta x}$ 存在

   记号及相关概念

   - $f(x) $ 在点 $x_0$ 处可导即$f(x) $ 在点 $x_0$ 处导数存在即 $f^{\prime}\left(x_{0}\right)=A$ (有限数)：
   
     导数：$f^{\prime}\left(x_{0}\right)=\displaystyle \lim _{\Delta x \rightarrow 0} \frac{\Delta y}{\Delta x}=\lim _{\Delta x \rightarrow 0} \frac{f\left(x_{0}+\Delta x\right)-f\left(x_{0}\right)}{\Delta x} $
   
     若令 $x = x_{0}+\Delta x$, 则 $f^{\prime}\left(x_{0}\right) = \displaystyle \lim _{x \rightarrow x_{0}} \frac{f(x)-f\left(x_{0}\right)}{x-x_{0}}$ 
   
   - 导数其他记法：$\left.\frac{\mathrm{d} y}{\mathrm{d} x}\right|_{x=x_{0}},\ \left.\frac{\mathrm{d} f(x)}{\mathrm{d} x}\right|_{x=x_{0}},\ \left.y^{\prime}\left(x_{0}\right),\ y^{\prime}\right|_{x=x_{0}}$ 
   
   - 单侧导数
   
     - 左导数：$f^{\prime}_-\left(x_{0}\right)=\displaystyle \lim _{\Delta x \rightarrow 0^-} \frac{\Delta y}{\Delta x}=\lim _{\Delta x \rightarrow 0^-} \frac{f\left(x_{0}+\Delta x\right)-f\left(x_{0}\right)}{\Delta x} $ 
     - 右导数：$f^{\prime}_+\left(x_{0}\right)=\displaystyle \lim _{\Delta x \rightarrow 0^+} \frac{\Delta y}{\Delta x}=\lim _{\Delta x \rightarrow 0^+} \frac{f\left(x_{0}+\Delta x\right)-f\left(x_{0}\right)}{\Delta x} $ 
     
   - 高阶导数：$f^{(n)}\left(x_{0}\right)= \displaystyle \lim _{\Delta x \rightarrow 0} \frac{f^{(n-1)}\left(x_{0}+\Delta x\right)-f^{(n-1)}\left(x_{0}\right)}{\Delta x},\ n \geqslant 2,\ n \in N $
   
3. 导数存在的判断

   - $f^{\prime}\left(x_{0}\right)=A$ $\iff$ $f^{\prime}_-\left(x_{0}\right) = f^{\prime}_+\left(x_{0}\right) = A$ 
   - $f^{\prime}\left(x_{0}\right)= \infty$ 是导数不存在的一种情况 
   
4. 导数的几何意义

   函数 $f(x) $ 在点 $x_0$ 处的导数就是曲线 $y = f(x)$ 在点 $(x_0,y_0)$ 处切线的斜率

## 微分

1. 本质

   以自变量增量为自变量的逼近原函数增量的一个线性函数

2. 定义

   组成：函数：$f(x),\ x \in I$，自变量：$x_0$，自变量增量：$\Delta x$ ，常数：$A$

   条件：存在与 $\Delta x$ 无关的常数 $A$ 使 $\Delta y=A \Delta x+o(\Delta x)$ ，其中 $\Delta y=f\left(x_{0}+\Delta x\right)-f\left(x_{0}\right)$ 

   记号及相关概念

   - $f(x) $ 在点 $x_0$ 处可微即$f(x) $ 在点 $x_0$ 处微分存在：$\left.\mathrm{d} y\right|_{x=x_{0}}=A \Delta x$ 或 $\mathrm{d}y = A \Delta x$ 
   - 微分其他记法：$\left.\mathrm{d} f(y)\right|_{x=x_{0}}=A \Delta x$ ，又 $\Delta x = \mathrm{d} x$，故 $\left.\mathrm{d} y\right|_{x=x_{0}}=A \mathrm{d} x$ 

3. 微分存在的判断

   - 写增量：$\Delta y=f\left(x_{0}+\Delta x\right)-f\left(x_{0}\right)$ ,写线性增量：$A \Delta x = f^\prime(x_0) \Delta x$ 

     求极限：$\displaystyle \lim _{\Delta x \rightarrow 0} \frac{\Delta y-A \Delta x}{\Delta x}$ ，若该极限为零，则 $f(x)$ 在 $x_0$ 处微分存在，否则不存在

   - $f(x)$ 在点 $x_0$ 处可微 $\iff$ $f(x)$ 在点 $x_0$ 处可导

4. 微分的几何意义

   函数 $f(x) $ 在点 $x_0$ 处可微，则在点 $(x_0,y_0)$ 附近可以用切线段近似代替曲线段

# 四则运算

## 导数

$[u(x) \pm v(x)]^{\prime}=u^{\prime}(x) \pm v^{\prime}(x)$ ，$[u(x) v(x)]^{\prime}=u^{\prime}(x) v(x)+u(x) v^{\prime}(x)$ 

$\left[\frac{u(x)}{v(x)}\right]^{\prime}=\frac{u^{\prime}(x) v(x)-u(x) v^{\prime}(x)}{[v(x)]^{2}}, v(x) \neq 0$ ，$[C u(x)]^{\prime}=C u^{\prime}(x)$ 

## 微分

$\mathrm{d}[u(x) \pm v(x)]=\mathrm{d} u(x) \pm \mathrm{d} v(x)$，$\mathrm{d}[u(x) v(x)]=u(x) \mathrm{d} v(x)+v(x) \mathrm{d} u(x)$ 

$\mathrm{d}\left[\frac{u(x)}{v(x)}\right]=\frac{v(x) \mathrm{d} u(x)-u(x) \mathrm{d} v(x)}{[v(x)]^{2}}, v(x) \neq 0$ ，$\mathrm d [C u(x)]=C \mathrm d u(x)$ 

# 特殊函数的求导

## 分段函数的求导

1. 设 $f(x)=\left\{\begin{array}{ll}f_{1}(x), & x \geqslant x_{0} \\ f_{2}(x), & x<x_{0}\end{array}\right.$ ，则 

   $f_{+}^{\prime}\left(x_{0}\right)=\displaystyle \lim _{x \rightarrow x_{0}^{+}} \frac{f_{1}(x)-f\left(x_{0}\right)}{x-x_{0}}, \ f_{-}^{\prime}\left(x_{0}\right)=\lim _{x \rightarrow x_{0}^{-}} \frac{f_{2}(x)-f\left(x_{0}\right)}{x-x_{0}}$ 

   根据左右导数来确定该点导数的情况

## 复合函数的求导

$\{f[g(x)]\}^{\prime}=f^{\prime}[g(x)] g^{\prime}(x)$，$\mathrm{d}\{f[g(x)]\}=f^{\prime}[g(x)] g^{\prime}(x) \mathrm{d} x$ (微分形式的不变性)

$\{f[g(x)]\}^{\prime}=\frac{\mathrm{d}\{f[g(x)]\}}{\mathrm{d} x}$，$f^{\prime}[g(x)]=\frac{\mathrm{d}\{f[g(x)]\}}{\mathrm{d} g(x)}$

## 反函数的求导

1. 设 $y=f(x)$ 可导，且导数不为零，若其存在反函数 $x=\varphi(y)$ ，则 $\varphi^{\prime}(y)=\frac{1}{f^{\prime}(x)}$ 

## 参数方程确定的函数的求导

1. 设函数 $y=y(x)$ 由参数方程 $\left\{\begin{array}{l}x=\varphi(t) \\ y=\psi(t)\end{array}\right.$ 确定，其中 $t$ 是参数，且 $\varphi(t)$，$\psi(t)$ 均对 $t$ 可导， $\varphi^\prime (t) \ne 0$ ，则 $\frac{\mathrm{d} y}{\mathrm{d} x}=\frac{\mathrm{d} y / \mathrm{d} t}{\mathrm{d} x / \mathrm{d} t}=\frac{\psi^{\prime}(t)}{\varphi^{\prime}(t)}$ 

# 函数求导常用方法

## 隐函数求导法

1. 设函数 $y=y(x)$ 是由方程 $F(x,y)=0$ 确定的可导函数，则在方程 $F(x,y)=0$ 两边对 $x$ 求导，其中  $y=y(x)$ ，所以最后可得到一个关于 $y^ \prime$ 的方程，解方程便可求出 $y^ \prime$ 

## 对数求导法

1. 设 $y=f(x)(f(x) > 0)$ ，等式两边取对数，得 $\ln y=\ln f(x)$ ，两边对 $x$ 求导，得 $\frac{1}{y} y^{\prime}=[\ln f(x)]^{\prime} \Rightarrow y^{\prime}=y[\ln f(x)]^{\prime}$ 

## 幂指函数求导法

1. 设函数 $u(x)^{v(x)}(u(x)>0, u(x) \neq 1)$ ，先化成指数函数 $u(x)^{v(x)}=\mathrm{e}^{v(x) \ln u(x)}$ ，再求导，得 $\left[u(x)^{v(x)}\right]^{\prime}=\left[\mathrm{e}^{v(x) \ln u(x)}\right]^{\prime}=u(x)^{v(x)}\left[{v}^{\prime}(x) \ln u(x)+v(x) \cdot \frac{u^{\prime}(x)}{u \cdot(x)}\right]$ 

# 函数的高阶导数

## 归纳法

1. 逐次求导，探索规律，得出通项

## 高阶求导公式

设 ${u}=u(x)$ ，$v=v(x)$ ，均 $n$ 阶可导，则

1. $[u \pm v]^{(n)}=u^{(n)} \pm v^{(n)}$ 

2. 莱布尼茨公式

   $\begin{aligned}(u v)^{(n)} &=u^{(n)} v+\mathrm{C}_{n}^{1} u^{(n-1)} v^{\prime}+\mathrm{C}_{n}^{2} u^{(n-2)} v^{\prime \prime}+\cdots+\mathrm{C}_{n}^{k} u^{(n-k)} v^{(k)}+\cdots+\mathrm{C}_{n}^{n-1} u^{\prime} v^{(n-1)}+u v^{(n)} \\ &=\sum_{k=0}^{n} C_{n}^{k} u^{(n-k)} v^{(k)} \end{aligned}$

## 泰勒公式

1. 先写出函数的泰勒公式，再通过比较系数来获得高阶导

   (1)任何一无穷阶可导函数(收敛)都可以写成 $y=f(x)=\displaystyle\sum_{n=0}^{\infty} \frac{f^{(n)}\left(x_{0}\right)}{n !}\left(x-x_{0}\right)^{n}$ 

   (2)通过已知公式将给定函数展开成幂级数，已知公式参见 《函数极限-泰勒展开式》

   (3)根据函数展开式的唯一性，比较(1)和(2)中公式的系数，就可得到 $f^{(n)}(x_0)$ 

## 参数方程确定的函数的二阶导数

1. 设函数 $y=y(x)$ 由参数方程 $\left\{\begin{array}{l}x=\varphi(t) \\ y=\psi(t)\end{array}\right.$ 确定，其中 $t$ 是参数，且 $\varphi(t)$，$\psi(t)$ 均对 $t$ 二阶可导，则 $\frac{\mathrm{d} y}{\mathrm{d} x}=\frac{\mathrm{d} y / \mathrm{d} t}{\mathrm{d} x / \mathrm{d} t}=\frac{\psi^{\prime}(t)}{\varphi^{\prime}(t)}$ ，$\frac{\mathrm{d}^{2} y}{\mathrm{d} x^{2}}=\frac{\mathrm{d}\left(\frac{\mathrm{d} y}{\mathrm{d} x}\right)}{\mathrm{d} x}=\frac{\mathrm{d}\left(\frac{\mathrm{d} y}{\mathrm{d} x}\right) / \mathrm{d} t}{\mathrm{d} x / \mathrm{d} t}=\frac{\psi^{\prime \prime}(t) \varphi^{\prime}(t)-\psi^{\prime}(t) \varphi^{\prime \prime}(t)}{\left[\varphi^{\prime}(t)\right]^{3}}$ 

## 反函数的二阶导数

1. 设函数 $y=f(x)$ 二阶可导，记 $f^{\prime}(x)=y_{x}^{\prime}$，$\varphi^{\prime}(y)=x_{y}^{\prime}\left(x_{y}^{\prime} \neq 0\right)$，则有

    $y_{x}^{\prime}=\frac{\mathrm{d} y}{\mathrm{d} x}=\frac{1}{\frac{\mathrm{d} x}{\mathrm{d} y}}=\frac{1}{x_{y}^{\prime}}$，$y_{x x}^{\prime \prime}=\frac{\mathrm{d}^{2} y}{\mathrm{d} x^{2}}=\frac{\mathrm{d}\left(\frac{\mathrm{d} y}{\mathrm{d} x}\right)}{\mathrm{d} x}=\frac{\mathrm{d}\left(\frac{1}{x_{y}^{\prime}}\right)}{\mathrm{d} x}=\frac{\mathrm{d}\left(\frac{1}{x_{y}^{\prime}}\right)}{\mathrm{d} y} \cdot \frac{1}{x_{y}^{\prime}}=\frac{-x_{yy}^{\prime \prime}}{\left(x_{y}^{\prime}\right)^{3}}$

   反过来，则有 $x_{y}^{\prime}=\frac{1}{y_{x}^{\prime}}$，$x_{y y}^{\prime \prime}=\frac{-y_{x x}^{\prime \prime}}{\left(y_{x}^{\prime}\right)^{3}}$

## 常见的高阶导数

$\left(a^{x}\right)^{(n)}=a^{x}(\ln a)^{n}(a>0, a \neq 1)$ ，$\left(\mathrm{e}^{x}\right)^{(n)}=\mathrm{e}^{x}$ ，$(\sin k x)^{(n)}=k^{n} \sin \left(k x+n \cdot \frac{\pi}{2}\right)$ ，$(\cos k x)^{(n)}=k^{n} \cos \left(k x+n \cdot \frac{\pi}{2}\right)$ ，$(\ln x)^{(n)}=(-1)^{n-1} \frac{(n-1) !}{x^{n}}(x>0)$ ，$[\ln (1+x)]^{(n)}=(-1)^{n-1} \frac{(n-1) !}{(1+x)^{n}}(x>-1)$ ，$\left[\left(x+x_{0}\right)^{m}\right]^{(n)}=m(m-1)(m-2) \cdots(m-n+1)\left(x+x_{0}\right)^{m-n}$ ，$\left(\frac{1}{x+a}\right)^{(n)}=\frac{(-1)^{n} n !}{(x+a)^{n+1}}$ 

# 变限积分的求导

1. 设 $F(x)=\int_{\varphi_1{(x)}}^{\varphi_{2}(x)} f(t) d t$ ，其中 $f(x)$ 在 $[a,b]$ 上连续，可导函数 $\varphi_1{(x)}$ 和 $\varphi_2{(x)}$ 的值域在 $[a,b]$ 上，则函数 $\varphi_1{(x)}$ 和 $\varphi_2{(x)}$ 的公共定义域上，有

   $F^{\prime}(x)=\frac{\mathrm{d}}{\mathrm{d} x}\left[\int_{\varphi_{1}(x)}^{\varphi_{2}{(x)}} f(t) \mathrm{d} t\right]=f\left[\varphi_{2}(x)\right] \varphi_{2}^{\prime}(x)-f\left[\varphi_{1}(x)\right] \varphi_{1}^{\prime}(x)$ 

# 基本求导公式

$\left(x^{a}\right)^{\prime}=a x^{a-1}(a为常数)$ ，$\left(a^{x}\right)^{\prime}=a^{x} \ln a(a>0, a \neq 1)$ ，$\left(\mathrm{e}^{x}\right)^{\prime}=\mathrm{e}^{x}$ ，$\left(\log _{\alpha} x\right)^{\prime}=\frac{1}{x \ln a}(a>0, a \neq 1)$ ，$(\ln x)^{\prime}=\frac{1}{x}$ ，$(\sin x)^{\prime}=\cos x$ ，$(\cos x)^{\prime}=-\sin x$ ，$(\arcsin x)^{\prime}=\frac{1}{\sqrt{1-x^{2}}}$ ，$(\arccos x)^{\prime}=-\frac{1}{\sqrt{1-x^{2}}}$ ，$(\tan x)^{\prime}=\sec ^{2} x$ ，$(\cot x)^{\prime}=-\csc ^{2} x$ ，$(\arctan x)^{\prime}=\frac{1}{1+x^{2}}$ ，$(\operatorname{arccot} x)^{\prime}=-\frac{1}{1+x^{2}}$ ，$(\sec x)^{\prime}=\sec x \tan x$ ，$(\csc x)^{\prime}=-\csc x \cot x$ ，$[\ln (x+\sqrt{x^{2}+1})]^{\prime}=\frac{1}{\sqrt{x^{2}+1}}$ ，$[\ln (x+\sqrt{x^{2}-1})]^{\prime}=\frac{1}{\sqrt{x^{2}-1}}$
